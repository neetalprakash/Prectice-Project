# MULTICOLLINEARITY WITH VARIANCE INFLATION FACTOR

```python
from scipy import stats
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
aba.sample()
```

```python
x=aba.iloc[:,:-1]
x.sample()
```

```python
y=aba.iloc[:,-1]
y.sample()
```

```python
def calc_vif(x):
    vif=pd.DataFrame()
    vif["variables"]=x.columns
    vif["VIF FACTOR"]=[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
    return(vif)
```

```python
calc_vif(x)
```

```python
aba=aba.drop("Length",axis=1)
```

```python
aba.head()
```

```python
calc_vif(aba)
```

```python
aba=aba.drop("Whole weight",axis=1)
```

# Spliting the data

```python
aba.head()
```

```python
aba.shape
```

```python
x=aba.iloc[:,:-1]
x.sample()
```

```python
y=aba.iloc[:,-1]
y.sample()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=43)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
predmnb=mnb.predict(x_test)
print(accuracy_score(y_test,predmnb))
print(confusion_matrix(y_test,predmnb))
print(classification_report(y_test,predmnb))
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
dtcpred=dtc.predict(x_test)
print(accuracy_score(y_test,dtcpred))
print(confusion_matrix(y_test,dtcpred))
print(classification_report(y_test,dtcpred))
```

```python
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
predknn=knn.predict(x_test)
print(accuracy_score(y_test,predknn))
print(confusion_matrix(y_test,predknn))
print(classification_report(y_test,predknn))
```

# Cross val score

```python
from sklearn.model_selection import cross_val_score
```

```python
score=cross_val_score(mnb,x,y,cv=5)
print(score)
print(score.mean())
print(score.std())
```

```python
score=cross_val_score(knn,x,y,cv=7)
print(score)
print(score.mean())
print(score.std())
```

```python
score=cross_val_score(dtc,x,y,cv=7)
print(score)
print(score.mean())
print(score.std())
```

# Grid search CV

```python
dtc=DecisionTreeClassifier(criterion='gini', splitter='best')
dtc.fit(x_train,y_train)
dtcpred=dtc.predict(x_test)
print(accuracy_score(y_test,dtcpred))
print(confusion_matrix(y_test,dtcpred))
print(classification_report(y_test,dtcpred))
```

```python
knn=KNeighborsClassifier(algorithm="brute",weights="uniform")
knn.fit(x_train,y_train)
predknn=knn.predict(x_test)
print(accuracy_score(y_test,predknn))
print(confusion_matrix(y_test,predknn))
print(classification_report(y_test,predknn))
```
