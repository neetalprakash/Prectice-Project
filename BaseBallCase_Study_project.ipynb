# Prediction

```python
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso

from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
bb.sample(3)
```

```python
x=bb.iloc[:,1:]
```

```python
x.head(2)
```

```python
y=bb.iloc[:,0]
```

```python
y.head(2)
```

```python
x.shape
```

```python
y.shape
```

# Spliting the data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=15)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
li=LinearRegression()
li.fit(x_train,y_train)
lipred=li.predict(x_test)

print('Mean absolute error:',mean_absolute_error(y_test,lipred))
print('Mean squared error:',mean_squared_error(y_test,lipred))
print('Root mean squraed Error:',np.sqrt(mean_squared_error(y_test,lipred)))
print(r2_score(y_test,lipred)*100)
```

```python
knn=KNeighborsRegressor()
knn.fit(x_train,y_train)
knnpred=knn.predict(x_test)
print('Mean absolute error:',mean_absolute_error(y_test,knnpred))
print('Mean squared error:',mean_squared_error(y_test,knnpred))
print('Root mean squraed Error:',np.sqrt(mean_squared_error(y_test,knnpred)))
print(r2_score(y_test,knnpred)*100)
```

```python
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
dtrpred=dtr.predict(x_test)
print(mean_absolute_error(y_test,dtrpred))
print(mean_squared_error(y_test,dtrpred))
print(np.sqrt(mean_squared_error(y_test,dtrpred)))
print(r2_score(y_test,dtrpred)*100)
```

```python
rd=Ridge()
rd.fit(x_train,y_train)
rdpred=rd.predict(x_test)
print(mean_absolute_error(y_test,rdpred))
print(mean_squared_error(y_test,rdpred))
print(np.sqrt(mean_squared_error(y_test,rdpred)))
print(r2_score(y_test,rdpred)*100)
```

```python
ls=Lasso()
ls.fit(x_train,y_train)
lspred=ls.predict(x_test)
print(mean_absolute_error(y_test,lspred))
print(mean_squared_error(y_test,lspred))
print(np.sqrt(mean_squared_error(y_test,lspred)))
print(r2_score(y_test,lspred)*100)
```

Observation:
    
By checking all algorithm Lasso model is best for the dataset.
