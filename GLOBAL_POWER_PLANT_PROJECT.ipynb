# MULTICOLLINEARITY

```python
from scipy import stats
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def vif(x):
    vif=pd.DataFrame()
    vif["variables"]=x.columns
    vif["VIF FACTOR"]=[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
    return(vif)
```

```python
vif(x)
```

# Prediction of capacity_mw

```python
from sklearn.model_selection import train_test_split
```

```python
li=LinearRegression()
li.fit(x_train,y_train)
lipred=li.predict(x_test)
print(mean_absolute_error(y_test,lipred))
print(mean_squared_error(y_test,lipred))
print(np.sqrt(mean_squared_error(y_test,lipred)))
print(r2_score(y_test,lipred)*100)
```

```python
ri=Ridge()
ri.fit(x_train,y_train)
ripred=ri.predict(x_test)
print(mean_absolute_error(y_test,ripred))
print(mean_squared_error(y_test,ripred))
print(np.sqrt(mean_squared_error(y_test,ripred)))
print(r2_score(y_test,ripred)*100)
```

```python
ml=MLPRegressor()
ml.fit(x_train,y_train)
mlpred=ml.predict(x_test)
print(mean_absolute_error(y_test,mlpred))
print(mean_squared_error(y_test,mlpred))
print(np.sqrt(mean_squared_error(y_test,mlpred)))
print(r2_score(y_test,mlpred)*100)
```

```python
ls=Lasso()
ls.fit(x_train,y_train)
lspred=ls.predict(x_test)
print(mean_absolute_error(y_test,lspred))
print(mean_squared_error(y_test,lspred))
print(np.sqrt(mean_squared_error(y_test,lspred)))
print(r2_score(y_test,lspred)*100)
```

```python
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
rfpred=rf.predict(x_test)
print(mean_absolute_error(y_test,rfpred))
print(mean_squared_error(y_test,rfpred))
print(np.sqrt(mean_squared_error(y_test,rfpred)))
print(r2_score(y_test,rfpred)*100)
```
